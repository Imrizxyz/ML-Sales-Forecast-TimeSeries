```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Monthly Sales Data (1 Year)
data = {
    "Month": pd.date_range(start="2023-01-01", periods=12, freq="M"),
    "Sales": [120, 135, 150, 160, 155, 170, 180, 190, 185, 200, 210, 220]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(8,4))
plt.plot(df["Month"], df["Sales"], marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Trend Over Time")
plt.grid()
plt.show()
```

```python
df["Moving_Avg"] = df["Sales"].rolling(window=3).mean()
print(df)
```

```python
last_3 = df["Sales"].tail(3)
forecast = last_3.mean()

print("Predicted Sales Next Month:", round(forecast))
```

```python
df["Month_Num"] = range(1, len(df)+1)
from sklearn.linear_model import LinearRegression

X = df[["Month_Num"]]
y = df["Sales"]

model = LinearRegression()
model.fit(X, y)

next_month = [[13]]
print("Trend-based Forecast:", int(model.predict(next_month)[0]))
```
